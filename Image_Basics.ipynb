## Image Basics

```python
import cv2
import pandas as pd
```

```python
path=r"/content/cute cat.jpg"
A=cv2.imread(path)
```

```python
A
```

```python
A[0]
```

```python
A.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(A)
# matplotlib showcases in BGR(Blue Green Red) format
# images are actually RGB(Red Green Blue) format
```

```python
B=cv2.cvtColor(A,cv2.COLOR_BGR2RGB)
plt.imshow(B)
```

```python
C=cv2.cvtColor(A,cv2.COLOR_BGR2GRAY)
plt.imshow(C,cmap="gray")
```
